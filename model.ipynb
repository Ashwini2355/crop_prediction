Potassium (K): Essential for enzyme activation, water regulation, and disease resistance. Types include available and total potassium.


Nitrogen (N): Critical for protein synthesis, chlorophyll production, and overall growth. Types include nitrate, ammonium, and total nitrogen.


Phosphorus (P): Vital for energy transfer and root development. Types include available and total phosphorus.


pH: Indicates soil acidity or alkalinity, affecting nutrient availability and microbial activity. Ranges from acidic to alkaline.


Temperature: Influences growth and development; includes air and soil temperature.

### Summary

- **Nitrogen (N):** Critical for protein synthesis and growth. Varies widely from 20 kg/ha (for legumes) to 200 kg/ha (for high-demand crops like sugarcane and maize).
- **Phosphorus (P):** Essential for energy transfer and root development. Typically ranges from 20 kg/ha (for pulses) to 80 kg/ha (for crops like sugarcane).
- **Potassium (K):** Important for water regulation and disease resistance. Ranges from 20 kg/ha (for some pulses) to 150 kg/ha (for crops like sugarcane and ginger).
- **pH:** Affects nutrient availability and soil health. Generally optimal between 5.5 and 7.5, with some crops having specific preferences.
- **Temperature:** Affects growth and development. Optimal temperatures typically range from 20°C to 35°C, depending on the crop.

These guidelines are general and can vary based on local soil conditions, climate, and specific crop varieties. Soil testing and tailored recommendations from local agricultural extensions or experts can provide more precise requirements.

Hyper- parameter tuning

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score

# Define the model with scaling
log_reg_pipeline = make_pipeline(StandardScaler(), LogisticRegression(max_iter=2000))

# Define the parameter grid
param_grid = {
    'logisticregression__C': [0.001, 0.01, 0.1, 1, 10, 100],
    'logisticregression__solver': ['liblinear', 'lbfgs']
}

# Initialize GridSearchCV
grid_search = GridSearchCV(log_reg_pipeline, param_grid, cv=5, scoring='accuracy')

# Fit GridSearchCV
grid_search.fit(X_train, y_train)

# Get the best parameters and the best model
best_log_reg_pipeline = grid_search.best_estimator_

# Make predictions
train_predictions = best_log_reg_pipeline.predict(X_train)
test_predictions = best_log_reg_pipeline.predict(X_test)

# Calculate accuracies
train_accuracy = accuracy_score(y_train, train_predictions)
test_accuracy = accuracy_score(y_test, test_predictions)

print(f"Logistic Regression with Scaling - Train Accuracy: {train_accuracy:.2f}, Test Accuracy: {test_accuracy:.2f}")
print(f"Best Parameters: {grid_search.best_params_}")
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.metrics import accuracy_score

# Define the model with scaling
svm_model = make_pipeline(StandardScaler(), SVC())

# Define the parameter grid
param_grid = {
    'svc__C': [0.1, 1, 10, 100],
    'svc__gamma': [0.001, 0.01, 0.1, 1],
    'svc__kernel': ['linear', 'rbf']  # Try different kernels
}

# Initialize GridSearchCV
grid_search = GridSearchCV(svm_model, param_grid, cv=5, scoring='accuracy')

# Fit GridSearchCV
grid_search.fit(X_train, y_train)

# Get the best parameters and the best model
best_svm = grid_search.best_estimator_

# Make predictions
train_predictions = best_svm.predict(X_train)
test_predictions = best_svm.predict(X_test)

# Calculate accuracies
train_accuracy = accuracy_score(y_train, train_predictions)
test_accuracy = accuracy_score(y_test, test_predictions)

print(f"SVM - Train Accuracy: {train_accuracy:.2f}, Test Accuracy: {test_accuracy:.2f}")
print(f"Best Parameters: {grid_search.best_params_}")
```
